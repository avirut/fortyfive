```
import pandas as pd
import numpy as np
```

```
arcos = pd.read_csv('/content/drive/My Drive/Global Health Project/Analysis/arcos_sanitized_am_200319.csv', index_col=[0,1], skipinitialspace=True)
arcos = arcos.rename(index={'Mcdowell': 'McDowell'})
wonder = pd.read_csv('/content/drive/My Drive/Global Health Project/Analysis/wonder_sanitized_am_200320.csv', index_col=[0,1], skipinitialspace=True)
cainc = pd.read_csv('/content/drive/My Drive/Global Health Project/Analysis/cainc_sanitized_am_200321.csv', index_col=[0,1], skipinitialspace=True)
bls = pd.read_csv('/content/drive/My Drive/Global Health Project/Analysis/bls_sanitized_am_200412.csv', index_col=[0,1], skipinitialspace=True)
```

```
# source: https://www.nrcs.usda.gov/wps/portal/nrcs/detail/wv/home/?cid=nrcs143_013697
# csv manually created from data above
fips = pd.read_csv('/content/drive/My Drive/Global Health Project/Analysis/wvfips.csv', index_col=[0], skipinitialspace=True)
```

```
joint = pd.concat([arcos, wonder, cainc, bls], axis=1)
```

```
for index, row in joint.iterrows():
  joint.at[index, 'FIPS'] = fips.loc[index[0], 'fips']
```

```
joint.to_csv('/content/drive/My Drive/Global Health Project/Analysis/joint_am_200412.csv')
```
