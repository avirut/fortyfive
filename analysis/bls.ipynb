```
import pandas as pd
import numpy as np
```

```
# src: https://www.bls.gov/lau/#cntyaa
yearrange = range(2006, 2014+1)
```

```
def getData(year, state):
  url = 'https://www.bls.gov/lau/laucnty' + str(year)[2:] + '.xlsx'
  bls = pd.read_excel(url, header=[4], skiprows=[5], 
                      names=['LAUS', 'StateFIPS', 'CountyFIPS', 'County', 'Year', 'empty', 'LaborForce', 'Employed', 'Unemployed', 'UnemployedRate'],
                      dtype={'StateFIPS': str, 'CountyFIPS': str, 'Year': str})
  bls.drop(['LAUS', 'empty'], axis=1, inplace=True)
  bls = bls.loc[bls['StateFIPS']==state]
  for index, row in bls.iterrows():
    name = row['County']
    bls.at[index, 'County'] = name[0:name.index(' ')]
  bls.reset_index(inplace=True)
  return bls
```

```
bls = getData(2014, '54')
counties = bls['County'].tolist()
years = [year for year in yearrange]*len(counties)
counties = [[county]*len(yearrange) for county in counties]
counties = [item for sublist in counties for item in sublist]
dIndex = pd.MultiIndex.from_arrays([counties, years], names=['County', 'Year'])
cols = ['UnemployedRate']
d = pd.DataFrame(np.zeros((len(counties), len(cols))), index=dIndex, columns=cols)
```

```
for year in yearrange:
  bls = getData(year, '54')
  for index, row in bls.iterrows():
    county = row['County']
    d.at[(county, year), 'UnemployedRate'] = row['UnemployedRate']
```

```
d.to_csv('/content/drive/My Drive/Global Health Project/Analysis/bls_sanitized_am_200412.csv')
```
